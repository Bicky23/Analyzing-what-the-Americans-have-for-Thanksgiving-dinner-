```python
# Part1 (Introduction to the dataset)
import pandas as pd
data = pd.read_csv("thanksgiving-2015-poll-data.csv", encoding = "Latin-1")
data.head()
# data.columns
```

```python
# Part 2 (Filtering out rows from a dataframe)
thanksgiving = data.iloc[:,1].value_counts()
print thanksgiving
```

```python
mask = data.iloc[:,1] == 'Yes'
people_who_celebrate_thanksgiving = data[mask]
people_who_celebrate_thanksgiving.head()
```

```python
# Part 3 (Using Value_counts To Explore Main Dishes)
people_who_celebrate_thanksgiving.iloc[:,2].value_counts()
```

```python
people_who_have_turkey = people_who_celebrate_thanksgiving.iloc[:,2] == 'Turkey'
celebrate_and_have_turkey = people_who_celebrate_thanksgiving[people_who_have_turkey]
celebrate_and_have_turkey['Do you typically have gravy?']
```
